Top 10 websites with Largest count of news articles

```python
website_counts = df['source_name'].value_counts()
```

```python
website_counts_df = website_counts.reset_index()
website_counts_df.columns = ['Website', 'Article Count']
```

```python
website_counts_df = website_counts_df.sort_values(by='Article Count', ascending=False)
```

```python
website_counts_df.head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(website_counts_df['Website'][:10], website_counts_df['Article Count'][:10], color='skyblue')
plt.title('Top 10 Websites with the Largest Counts of News Articles')
plt.xlabel('Website')
plt.ylabel('Article Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
